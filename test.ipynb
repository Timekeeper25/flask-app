```python
import pandas as pd
from datetime import datetime
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
data_csv = pd.read_csv('quesioner.csv', sep=',')

data_csv
```

## Melihat Nilai Kategorikal Yang Diubah Oleh LabelEncoder

#### Sesuai urutan list dimulai dari 0, baru kemudian dibuat mapnya di app.py

```python
label_encoders = {}

columns_to_encode = ['L/P','Minat', 'Bakat', 'Hobi']

for col in columns_to_encode:
    le = LabelEncoder()
    data_csv[col] = le.fit_transform(data_csv[col]) + 1
    label_encoders[col] = le

# Melihat nilai-nilai kategorikal yang di-label-encode untuk setiap kolom
for col in columns_to_encode:
    print(f"Nilai-nilai kategorikal yang di-label-encode untuk kolom '{col}': {label_encoders[col].classes_} \n")

data_csv
```
